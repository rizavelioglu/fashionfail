```python
import pandas as pd
import json 
import os
import matplotlib.pyplot as plt
```

```python
AD_FILEPATH = "/home/sinanharms/Dokumente/shopdata/adidas/"
adidas_json = [AD_FILEPATH + pos for pos in os.listdir(AD_FILEPATH) if pos.endswith(".json")]
dflist = []
for file in adidas_json:
    df = pd.read_json(file, lines=True)
    dflist.append(df)
    
df_adidas = pd.concat(dflist, axis=0).reset_index()
```

```python
filtered = df_adidas[(df_adidas.category != "Kleidung") & (df_adidas.category != "Schuhe")]
filtered.category.value_counts().plot(kind="bar")
```

```python
filtered.category.unique()
```

```python
filtered.subcategory.value_counts().plot(kind="bar")
```
